```python
import pandas as pd
import plotnine 
from plotnine import *

my_file = pd.read_table('gapminder.txt')
```

```python
my_file
```

```python
(ggplot(my_file) + aes(x='gdpPercap', y='lifeExp') + geom_point())
```

### Make a plot using ggplot where x = pop and y = lifeExp

```python
(ggplot(my_file) + aes(x='pop', y='lifeExp') + geom_point())
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp') + geom_point())
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp', color='continent') + geom_point())
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp', color='continent', by='country') + geom_line())
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp', color='continent') + geom_line())
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp', color='continent') + 
 geom_point() + geom_line() )
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp') + 
  geom_line(aes(color='continent')) + geom_point() )
```

```python
(ggplot(my_file) + aes(x='year', y='lifeExp', by='country') + 
  geom_line(aes(color='continent')) + geom_point() )
```

```python
countries = my_file[my_file.country.str.startswith('A') |
                    my_file.country.str.startswith('Z')]
```

```python
countries
```

```python
(ggplot(countries) + aes(x='year', y='lifeExp', color='continent') + 
  geom_line(countries) +
 facet_wrap('country') )
```

```python
p = (ggplot(countries) + aes(x='gdpPercap', y='lifeExp', size='pop', 
                         color='continent') + 
  geom_point() )
```

```python
p
```

```python
ggsave(p, filename = 'jees.png')
```
